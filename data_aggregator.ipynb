# Модуль 2. Универсальный агрегатор данных

```python
def aggregate_values(data, threshold=0):
    numbers = []
    values = list(data.values()) if isinstance(data, dict) else data
    
    for item in values:
        if isinstance(item, (int, float)):
            if item >= threshold:
                numbers.append(item)
        elif isinstance(item, (list, tuple)):
            for subitem in item:
                if isinstance(subitem, (int, float)) and subitem >= threshold:
                    numbers.append(subitem)
    
    total = sum(numbers)
    count = len(numbers)
    avg = round(total / count, 2) if count > 0 else 0
    return total, avg, count
```

```python
# Тест 1. Смешанный словарь
test_dict = {
    'a': 1500,
    'b': [200, 300, 400],
    'c': 'строка',
    'd': 800,
    'f': (50, 60, 'текст')
}

total, avg, cnt = aggregate_values(test_dict)
print(f'Сумма: {total}, Среднее: {avg}, Чисел: {cnt}')
```

```python
# Тест 2. С порогом
total, avg, cnt = aggregate_values(test_dict, threshold=500)
print(f'Сумма (>=500): {total}, Среднее: {avg}, Чисел: {cnt}')
```

```python
# Тест 3. Список с вложением
test_list = [100, [200, 300, 'x'], (400, 500), 'abc', 600]
total, avg, cnt = aggregate_values(test_list)
print(f'Сумма: {total}, Среднее: {avg}, Чисел: {cnt}')
```
